```python
#1st
import numpy as np

# Mean score
mean_score = 85

# No. of learners
num_learners = 10

# variance is zero, all scores are identical to the mean score
scores = np.full(num_learners, mean_score)

# Print the scores obtained by all learners
print("Scores obtained by all learners:", scores)

# Descriptive statistics
mean = np.mean(scores)
variance = np.var(scores)
std_dev = np.std(scores)
min_score = np.min(scores)
max_score = np.max(scores)
range_scores = np.ptp(scores)  # Peak-to-peak (range) of scores
median_score = np.median(scores)

# Print descriptive statistics
print("\nDescriptive Statistics:")
print("Mean score:", mean)
print("Variance of scores:", variance)
print("Standard deviation of scores:", std_dev)
print("Minimum score:", min_score)
print("Maximum score:", max_score)
print("Range of scores:", range_scores)
print("Median score:", median_score)
```

```python
#18. Do the higher-income holders spend more on credit cards?
import statsmodels.api as sm

# Scatter Plot with Regression Line
plt.figure(figsize=(12, 6))
sns.regplot(x='Income', y='CCAvg', data=df, scatter_kws={'s':50}, line_kws={'color':'red'})
plt.xlabel('Income')
plt.ylabel('Average Credit Card Spend (CCAvg)')
plt.title('Relationship between Income and Credit Card Spending')
plt.show()

# Pearson Correlation Coefficient
correlation = df['Income'].corr(df['CCAvg'])
print(f"Pearson correlation coefficient between Income and CCAvg: {correlation:.2f}")

# Linear Regression Analysis
X = df['Income']
y = df['CCAvg']
X = sm.add_constant(X)  # Add constant for intercept

model = sm.OLS(y, X).fit()  # Fit the model

#summary of the regression model
print("\nLinear Regression Model Summary:")
print(model.summary())
```

```python
#19. How many customers use online banking? Do customers using bank internet facilities have higher incomes?
import scipy.stats as stats
# 1. Counting Number of Customers Using Online Banking
# Assume 'Online' column indicates usage with 1 for users and 0 for non-users
num_online_banking_users = df[df['Online'] == 1].shape[0]
print(f"Number of customers using online banking: {num_online_banking_users}")

# Average Income for Both Groups
# Average income for customers using internet banking
average_income_online = df[df['Online'] == 1]['Income'].mean()

# Average income for customers not using internet banking
average_income_no_online = df[df['Online'] == 0]['Income'].mean()

print(f"Average income of customers using online banking: {average_income_online:.2f}")
print(f"Average income of customers not using online banking: {average_income_no_online:.2f}")

# 3. t-test to Check if the Income Difference is Statistically Significant
# Drop missing values in 'Income' column
income_online = df[df['Online'] == 1]['Income'].dropna()
income_no_online = df[df['Online'] == 0]['Income'].dropna()

#  t-test
t_stat, p_value = stats.ttest_ind(income_online, income_no_online)

print(f"\nT-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.2f}")

#  p-value
if p_value < 0.05:
    print("The difference in income between customers using and not using online banking is statistically significant.")
else:
    print("The difference in income is not statistically significant.")
```

```python
###20. Using the z-score of the income variable, find out the number of observations outside the +-3σ.

df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
print(df['Income'].describe())
df = df.dropna(subset=['Income'])

# Mean and standard deviation of the 'Income' column
mean_income = df['Income'].mean()
std_income = df['Income'].std()

print(f"Mean Income: {mean_income:.2f}")
print(f"Standard Deviation of Income: {std_income:.2f}")

# Z-scores for the 'Income' column
df['ZScore_Income'] = (df['Income'] - mean_income) / std_income

# number of observations with Z-scores outside the ±3 range
num_outside_3sigma = df[(df['ZScore_Income'] < -3) | (df['ZScore_Income'] > 3)].shape[0]

print(f"Number of observations outside ±3σ: {num_outside_3sigma}")
```
